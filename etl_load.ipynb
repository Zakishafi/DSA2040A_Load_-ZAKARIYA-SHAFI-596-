```python
import pandas as pd

df_full = pd.read_csv("transformed_full.csv")
df_inc = pd.read_csv("transformed_incremental.csv")
```

```python
import sqlite3
from pathlib import Path

# Create 'loaded' folder
Path("loaded").mkdir(parents=True, exist_ok=True)

# Connect to SQLite
conn = sqlite3.connect("loaded/full_data.db")

# Load data
df_full.to_sql("full_data", conn, if_exists="replace", index=False)
df_inc.to_sql("incremental_data", conn, if_exists="replace", index=False)

# Verify the load
preview = pd.read_sql("SELECT * FROM full_data LIMIT 5", conn)
display(preview)

conn.close()
```
